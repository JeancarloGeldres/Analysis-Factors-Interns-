```python
#Importamos todos los paquetes que vamos a necesitar
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from math import sqrt
```

```python
# cargamos nuestro dataset
modelo = pd.read_csv('fourth_banks.csv',sep=',')
```

```python
modelo_bcp = modelo[(modelo['banco']=='BCP')]
modelo_bbva = modelo[(modelo['banco']=='Banco Continental')]
modelo_interbank = modelo[(modelo['banco']=='Interbank')]
modelo_scotiabank = modelo[(modelo['banco']=='Scotiabank Peru')]

modelo_roa_bcp = modelo_bcp.groupby(['year', 'banco']).sum().reset_index()
modelo_roa_bbva = modelo_bbva.groupby(['year', 'banco']).sum().reset_index()
modelo_roa_interbank = modelo_interbank.groupby(['year', 'banco']).sum().reset_index()
modelo_roa_scotiabank = modelo_scotiabank.groupby(['year', 'banco']).sum().reset_index()

modelo_bcp2 = modelo_roa_bcp[modelo_roa_bcp['banco'] == 'BCP']
modelo_bbva2 = modelo_roa_bbva[modelo_roa_bbva['banco'] == 'Banco Continental']
modelo_interbank2 = modelo_roa_interbank[modelo_roa_interbank['banco'] == 'Interbank']
modelo_scotiabank2 = modelo_roa_scotiabank[modelo_roa_scotiabank['banco'] == 'Scotiabank Peru']

plt.plot(modelo_bcp2['year'], modelo_bcp2['roa'], label='roa_bcp')
plt.plot(modelo_bcp2['year'], modelo_bcp2['pbi'], label='pbi')
plt.plot(modelo_bbva2['year'], modelo_bbva2['roa'], label='roa_bbva')
plt.plot(modelo_interbank2['year'], modelo_interbank2['roa'], label='roa_in')
plt.plot(modelo_scotiabank2['year'], modelo_scotiabank2['roa'], label='roa_sc')

plt.xlabel('AÃ±os')
plt.ylabel('Tasa de crecimiento (%)')
plt.legend()
plt.xticks(rotation = 45)
plt.show(100)
```

```python
# cargamos las funciones necesarias
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
model = LinearRegression().fit(X_train, y_train)
model
```

```python
# DiagnÃ³stico errores (residuos) de las predicciones de entrenamiento
# ==============================================================================
y_train = y_train.values.flatten()
prediccion_train = results1.predict(exog = X_train)
residuos_train   = prediccion_train - y_train
```

```python
# GrÃ¡ficos
# ==============================================================================
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(9, 8))

axes[0, 0].scatter(y_train, prediccion_train, edgecolors=(0, 0, 0), alpha = 0.4)
axes[0, 0].plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()],
                'k--', color = 'black', lw=2)
axes[0, 0].set_title('Valor predicho vs valor real', fontsize = 10, fontweight = "bold")
axes[0, 0].set_xlabel('Real')
axes[0, 0].set_ylabel('PredicciÃ³n')
axes[0, 0].tick_params(labelsize = 7)

axes[0, 1].scatter(list(range(len(y_train))), residuos_train,
                   edgecolors=(0, 0, 0), alpha = 0.4)
axes[0, 1].axhline(y = 0, linestyle = '--', color = 'black', lw=2)
axes[0, 1].set_title('Residuos del modelo', fontsize = 10, fontweight = "bold")
axes[0, 1].set_xlabel('id')
axes[0, 1].set_ylabel('Residuo')
axes[0, 1].tick_params(labelsize = 7)

sns.histplot(
    data    = residuos_train,
    stat    = "density",
    kde     = True,
    line_kws= {'linewidth': 1},
    color   = "firebrick",
    alpha   = 0.3,
    ax      = axes[1, 0]
)

axes[1, 0].set_title('DistribuciÃ³n residuos del modelo', fontsize = 10,
                     fontweight = "bold")
axes[1, 0].set_xlabel("Residuo")
axes[1, 0].tick_params(labelsize = 7)


sm.qqplot(
    residuos_train,
    fit   = True,
    line  = 'q',
    ax    = axes[1, 1], 
    color = 'firebrick',
    alpha = 0.4,
    lw    = 2
)
axes[1, 1].set_title('Q-Q residuos del modelo', fontsize = 10, fontweight = "bold")
axes[1, 1].tick_params(labelsize = 7)

axes[2, 0].scatter(prediccion_train, residuos_train,
                   edgecolors=(0, 0, 0), alpha = 0.4)
axes[2, 0].axhline(y = 0, linestyle = '--', color = 'black', lw=2)
axes[2, 0].set_title('Residuos del modelo vs predicciÃ³n', fontsize = 10, fontweight = "bold")
axes[2, 0].set_xlabel('PredicciÃ³n')
axes[2, 0].set_ylabel('Residuo')
axes[2, 0].tick_params(labelsize = 7)

# Se eliminan los axes vacÃ­os
fig.delaxes(axes[2,1])

fig.tight_layout()
plt.subplots_adjust(top=0.9)
fig.suptitle('DiagnÃ³stico residuos', fontsize = 12, fontweight = "bold");
```
